```python
def center_sq_view(arr):
    w = arr.shape[1]
    h = arr.shape[0]
    mid = w / 2
    return arr[:, mid - h / 2 : mid + h / 2]
```

```python
def window_sq(arr):
    w = arr.shape[1]
    h = arr.shape[0]
    mid = w / 2
    return (arr[:, :h], arr[:, h:])
```
